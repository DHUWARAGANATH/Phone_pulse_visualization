```python
#agg_transaction

file1="C:/Users/dhuwa/Downloads/project folder/phonepe/pulse/data/aggregated/transaction/country/india/state"
agg_tran_state=os.listdir(file1)

columns1 ={'states':[],'years':[],'quarters':[],'Transaction_type':[],'Transaction_count':[],'Transaction_amount':[]}

for state in agg_tran_state:
    current_states=os.path.join(file1,state)
    agg_year_list=os.listdir(current_states)
    

    for year in agg_year_list:
        current_year=os.path.join(current_states,year)
        agg_file_list=os.listdir(current_year)
    
        for file in agg_file_list:
            current_file=os.path.join(current_year,file)
            data=open(current_file)
    
            f=json.load(data)
            for i in f['data']['transactionData']:
                name=i['name']
                count=i['paymentInstruments'][0]['count']
                amount=i['paymentInstruments'][0]['amount']
                columns1['Transaction_type'].append(name)
                columns1['Transaction_count'].append(count)
                columns1['Transaction_amount'].append(amount)
                columns1['states'].append(state)
                columns1['years'].append(year)
                columns1['quarters'].append(int(file.strip('.json')))

agg_transaction =pd.DataFrame(columns1)
agg_transaction['states']=agg_transaction['states'].str.replace('-',' ')
agg_transaction['states']=agg_transaction['states'].str.title()
agg_transaction['states']=agg_transaction['states'].str.replace('Dadra & Nagar Haveli & Daman & Diu',' Dadra and Nagar Haveli and Daman and Diu')
```

```python
#agg_user

file2 = "C:/Users/dhuwa/Downloads/project folder/phonepe/pulse/data/aggregated/user/country/india/state"
agg_user_state=os.listdir(file2)

columns2 ={'states':[],'years':[],'quarters':[],'Brand':[],'Transaction_count':[],'Percentage':[]}

for state in agg_user_state:
    current_states = os.path.join(file2,state)
    agg_user_year = os.listdir(current_states)
    
    for year in agg_user_year:
        current_year = os.path.join(current_states,year)
        agg_user_file = os.listdir(current_year)
        
        for file in agg_user_file:
            current_file = os.path.join(current_year,file)
            data = open(current_file)
            
            f2 = json.load(data)
            
            try:
                for i in f2['data']['usersByDevice']:
                    brand = i['brand']
                    count = i['count']
                    percentage = i['percentage']
                    columns2['Brand'].append(brand)
                    columns2['Transaction_count'].append(count)
                    columns2['Percentage'].append(percentage)
                    columns2['states'].append(state)
                    columns2['years'].append(year)
                    columns2['quarters'].append(int(file.strip('.json')))
        
            except:
                pass

agg_user =pd.DataFrame(columns2)
agg_user['states']=agg_user['states'].str.replace('-',' ')
agg_user['states']=agg_user['states'].str.title()
agg_user['states']=agg_user['states'].str.replace('Dadra & Nagar Haveli & Daman & Diu',' Dadra and Nagar Haveli and Daman and Diu')
agg_user['Percentage'] = (agg_user['Percentage'] * 100).round().astype(int)
```
